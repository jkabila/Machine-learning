```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.read_csv("/content/diabetes_prediction_dataset.csv")
df
```

```python
encode=OrdinalEncoder()
df.gender=encode.fit_transform(df[["gender"]])
```

```python
df.gender.unique()
```

```python
encode=OrdinalEncoder()
df.smoking_history=encode.fit_transform(df[["smoking_history"]])
```

```python
encode=OrdinalEncoder()
df.diabetes=encode.fit_transform(df[["diabetes"]])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop("diabetes",axis=1)
y=df["diabetes"] # dependent
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
x_train.shape,x_test.shape   #no of rows,columns
```

```python
y_train.shape,y_test.shape   # 75000 rows,1 column,25000 rows,1 column
```

```python
model=DecisionTreeClassifier().fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
# entropy
model=DecisionTreeClassifier(criterion="entropy").fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```
