```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/bitcoin.csv")
df.head()
```

```python
df.drop(['Date'],axis=1,inplace=True)
```

```python
predictiondays=30
df["Prediction"]=df[["Price"]].shift(-predictiondays)
df.head()
```

```python
df.tail()
```

```python
x=np.array(df.drop(["Prediction"],axis=1))
x=x[:len(df)-predictiondays]
```

```python
print(x)
```

```python
#creating dependent dataset
#converting tha data from into numpy array
y=np.array(df["Prediction"])
y=y[:-predictiondays]
print(y)
```

```python
#spliting the data set into test and train data,test=20%,train=80%
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
predictiondays_array=np.array(df.drop(["Prediction"],axis=1))[-predictiondays:]
print(predictiondays_array)
```

```python
#Machine learning
from sklearn.svm import SVR
svr_rbf=SVR(kernel="rbf",C=1e3,gamma=0.00001)
svr_rbf.fit(x_train,y_train)
```

```python
SVR(C=1000.0, cache_size=200, coef0=0.0, degree=3, epsilon=0.1, gamma=1e-05,
    kernel='rbf', max_iter=-1, shrinking=True, tol=0.001, verbose=False)
```

```python
#Testing the data
svr_rbf_confidence = svr_rbf.score(x_test,y_test)
print('SVR_RBF accuracy :',svr_rbf_confidence)
```

```python
svm_prediction = svr_rbf.predict(x_test)
print(svm_prediction)
print()
print(y_test)
```
