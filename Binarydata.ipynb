```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
import plotly.express as px
```

```python
df=pd.read_csv("/content/Binary predictors.csv")
df
```

```python
df["Admitted"]=df["Admitted"].replace({"Yes":1,"No":0})
```

```python
df
```

```python
df["Gender"]=df["Gender"].replace({"Male":1,"Female":0})
```

```python
df
```

```python
#admitted-dependent
#mark,gender-independent
```

```python
y=df["Admitted"]
x1=df["Marks"]
```

```python
fig=px.scatter(df,x="Marks",y="Admitted")
fig.show()
```

```python
 x=sm.add_constant(x1)
```

```python
result=sm.OLS(y,x).fit()
```

```python
yhat=result.params[0]+result.params[1]*x1
```

```python
result.summary()
```

```python
fig = px.scatter(df, x="Marks", y="Admitted")
fig2=px(x1,yhat,color="red")
fig2.show()
```
